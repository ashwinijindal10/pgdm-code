```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

x= np.linspace(5,100,100)
y= np.linspace(2,44,100)

plt.plot(x,y,'b+', x*2,y/2,'ro')
plt.show()
```

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

x=np.linspace(3,100,100)
y=np.log(x)

plt.figure(1)

plt.subplot(121)
plt.title('y=log(x)')
plt.plot(x,y)


plt.subplot(122)
plt.title('y=log(x)**2')
plt.plot(x,y**2)

plt.show()


```
