```python
from nltk.tokenize import word_tokenize,sent_tokenize,regexp_tokenize
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

stemmer = PorterStemmer()
lemmalizer = WordNetLemmatizer()

def preprocess(doc, stem =True): 
    doc= doc.lower()   
    words= word_tokenize(doc)
    words = [word for word in words if word not in stopwords.words('english')]
    
    if stem :
         words = [ stemmer.stem(word) for word in words ]
    else :
         words = [ lemmalizer.lemmatize(word,pos='v') for word in words]
        
    doc= ' '.join(words)
    return doc


```

```python
import pandas as pd 
pd.set_option('max_colwidth',1000)

spam_data = pd.read_csv('SMSSpamCollection.txt',sep='\t' ,names=['label','message'])


spam_data.head()
```

```python
spam = spam_data.iloc[:50,:]

```

```python
#bag of words 
from sklearn.feature_extraction.text import CountVectorizer

messages = [preprocess(m,False)for m in spam.message]

vectorize= CountVectorizer()
model = vectorize.fit_transform(messages)

print(sum(sum(model.toarray())))
print(model.shape)
messages
```
